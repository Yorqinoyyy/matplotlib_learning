### 4.1 Chiziqli grafiklar. `matplotlib` paketi.


```python
import pandas as pd
import numpy as np
```

Data Science muhandisi uchun ma'lumotlarni grafik ko'rinishida taqdim qila bilish muhim ko'nikmalardan biri hisoblanadi. Ayniqsa, oddiy odamlar (buyurtmachi, rahbar, hamkasblar) uchun sonlar va jadvallardan ko'ra grafiklarni tushunish osonroq kechadi.

Muhandis sifatida, ba'zida o'zimiz uchun ham grafiklardan qo'shimcha ma'lumotlar olishimiz mumkin. Grafiklar nafaqat ma'lumotlarga ishlov berish balki yakuniy natijalarni taqdim qilishda ham faol ishlatiladi.

Pythonda grafiklar bilan ishlash uchun eng mashhur kutubxonalar bu matplotlib va u asosida qurilgan seaborn kutubxonalaridir.

Ushbu darsimizda matplotlib bilan tanishamiz.



```python
import  matplotlib.pyplot as plt
%matplotlib inline
```

Quyida biz 0 dan 100 gacha sonlarni yaratib olamiz va to'g'ri chiziq chizamiz:

```python
sonlar = np.arange(1, 101)
plt.plot(sonlar)
plt.show()
```

```python
x = np.arange(1, 101)
y = x**2
plt.plot(x, y)
plt.show()
```

  sin(x)  funksiyasining  0  va  5π  oralig'idagi grafigini chizamiz.

```python
x = np.arange(0, 5 * np.pi, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x,y, linewidth=1.5, linestyle='-.',color='#A433A6', alpha=0.9, marker='*')
plt.show()
```

Grafik hajmini o'zgartirish va grafik haqida qo'shimcha ma'lumotlar qo'shish uchun ham maxsus funksiyalar bor

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y,
         linewidth=1.5,
         linestyle='--',
         color='#A433A6',
         alpha=0.6,
         marker='h')
plt.title('sin(x) funksiya grafigi')
plt.xlabel("pi qiymatlari")
plt.ylabel("sin(x) qiymatlari")
plt.grid(color='k', alpha=0.5, linestyle='--')
plt.show()
```

2 ta grafik chizish

```python
x = np.arange(0, 5 * np.pi, 0.1)
y = np.arange(0, 4 * np.pi, 0.1)
sin = np.sin(x)
cos = np.cos(y)
plt.figure(figsize=(12, 6))
plt.plot(x, sin,
         linewidth=1.4,
         linestyle='-.',
         color="#A433A6",
         alpha=0.8,
         marker='h')

plt.plot(y, cos,
         linewidth=1.4,
         linestyle=':',
         color="c",
         alpha=0.8,
         marker='v')
plt.title('sin(x) va cos(y) funksiyalar grafigi')
plt.xlabel("pi qiymatlari")
plt.ylabel("sin(x) va cos(y) qiymatlari")
plt.legend(['sin(x)', 'cos(y)'], loc=4, fontsize='x-large')
plt.grid(color='tomato', alpha=0.5, linestyle='--')
plt.savefig('sin-cos_grafik.png')
```
