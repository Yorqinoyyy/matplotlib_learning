## 1-VAZIFA: Sahifadagi jadvalni pandas dataframega aylantirish va:

```python
# JAVOBNI SHUYERGA YOZING
url= "https://en.tutiempo.net/climate/ws-384570.html"
tables = pd.read_html(url, index_col=0)
tables[0].tail(5)
df = tables[0][tables[0].index>1990]
df.head()
```

### **SAVOL**: Yuqoridagi natijadan qanday xulosaga kelish mumkin? DF dagi sonlar bilan ishlasak bo'ladimi?

```python
# JAVOB:
# Afsuski, jadvalda mavjud bo'lmagan qiymatlar o'rniga tire (-) belgisi qo'yib ketilgan,
# shu sababdan jadval ustunlari son (int, float) emas matn (object) bo'lib qolgan.
```

## 4-VAZIFA. `df`dagi barcha `-` belgilarni `NaN` qiymat bilan almashtirish.

```python
# JAVOBNI SHUYERGA YOZING
df = df.replace("-", np.nan)
df
```

## 5-VAZIFA. `df` ustunlari qiymatlarini matndan o'nlik songa (`float`) o'zgartirish.

```python
# JAVOBNI SHUYERGA YOZING
df = df.astype(dtype=float)
df.head(10)
```

## 6-VAZIFA. `NaN` qiymatlar bilan ishlaymiz
### 6.1. Har bir ustundagi `NaN` qiymatlar soni va ularning proprosiyasini topish.

```python
# JAVOBNI SHUYERGA YOZING
nan_counts = df.isnull().sum()
print("Ustundagi NaN qiymatlar soni:\n", nan_counts)
nan_ratio = df.isnull().sum()/len(df)*100
print("\nProporsiya:\n", nan_ratio)
```

### 6.2. Eng ko'p `nan` qiymat mavjud ustunni tashlab yuborish va natijani saqlab qolish.

```python
# JAVOBNI SHUYERGA YOZING
df = df.drop(columns=[nan_counts.idxmax()])
df
```

### 6.3. Qolgan `NaN` qiymatlarni o'zingiz to'g'ri deb bilgan usulda to'ldiring (tashlab yubormang) va natijani saqlab qoling.

```python
# JAVOBNI SHUYERGA YOZING
df.fillna(method='ffill', inplace=True)
df
```

### 6.4. Yakuniy `df` ni csv faylga saqlab, kompyuterga (yoki githubga) yuklab olish

```python
# JAVOBNI SHUYERGA YOZING
df.to_csv("matplotlib_cleaned_data1.csv", index=False)
```

##7-VAZIFA. Chiziqli grafiklar

### 7.1. `df`dan yillik o'rtacha haroratni ko'rsatuvchi ustunni `x` o'zgaruvchiga saqlash va `plt.plot()` yordamida chiziqli grafik chizish.

```python
# JAVOBNI SHUYERGA YOZING
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x= df['T']
plt.plot(x)
plt.show()
```

### 7.2. Yuqoridagi grafikni hajmini kattalashtirish, chiziq qalingligi, turi, rangi, markerlar turini istalgancha o'zgartirish

```python
# JAVOBNI SHUYERGA YOZING
plt.figure(figsize=(12,6))
plt.plot(x, linewidth=1.6,
         linestyle='-.',
         color='#8E44AD',
         alpha=0.9,
         marker='o')
plt.show()
```

### 7.3. Yuqoridagi grafikda x va y o'qlarini nomlash, grafikka ham umumiy nom berish

```python
# JAVOBNI SHUYERGA YOZING
plt.figure(figsize=(12,6))
plt.plot(x, linewidth=1.6,
         linestyle='-.',
         color='#8E44AD',
         alpha=0.9,
         marker='o')
plt.title("Toshkent shahri uchun yillik o'rtacha harorat")
plt.xlabel("Yil")
plt.ylabel("Harorat")
plt.grid(color='b', alpha=0.5, linestyle='--')
plt.show()
```

## 8-VAZIFA. Toshkent shahri uchun, har bir yilda o'rtacha min va max haroratlarni ko'rsatuvchi grafik chizish. **Grafikni faylga saqlab olish**

```python
# JAVOBNI SHUYERGA YOZING
minT = df['Tm']
maxT = df['TM']
plt.figure(figsize=(16,8))
plt.plot(minT,
         linewidth=1.8,
         linestyle='-.',
         color='#8E44AD',
         alpha=1,
         marker='*',
         markerfacecolor='g',
         label="O'rtacha minimum harorat")
plt.plot(maxT,
         linewidth=1.8,
         linestyle='-.',
         color='#E63115',
         alpha=1,
         marker='o',
         markerfacecolor='r',
         label="O'rtacha maximum harorat")
plt.title("Toshkent shahri uchun yillik o'rtacha min va max harorat")
plt.xlabel("Yil")
plt.ylabel("Harorat")
plt.legend(['Tm','TM'], loc=0)
plt.grid(color='tomato', alpha=0.6, linestyle='--')
plt.savefig('tempMinMax.png')
plt.show()
```
